```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/train.csv")
```

```python
column_names = ['Age', 'Fare']
summary_stats = df[column_names].describe()
```

```python
summary_stats
```

```python

class_counts = df['Pclass'].value_counts()
class_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.title('Count of Passengers in Each Passenger Class')
plt.xticks(rotation=0)
plt.show()
```

```python
df['Age'].plot(kind='hist', bins=20, color='salmon')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Distribution of Passenger Ages')
plt.show()
```

```python

gender_counts = df['Sex'].value_counts()
gender_counts.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
df['Fare'].plot(kind='box', vert=False, color='gold')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Pclass', y='Age', data=df, palette='Set3')
plt.title('Age Distribution by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Survived', hue='Pclass', data=df, palette='Set2')
plt.title('Survivors by Passenger Class')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.legend(title='Class', loc='upper right', labels=['1st', '2nd', '3rd'])
plt.show()
```

```python
numerical_features = ['Age', 'Fare']
sns.pairplot(df, vars=numerical_features, hue='Pclass', palette='Set3')
plt.suptitle('Pair Plot of Numerical Features by Passenger Class')
plt.show()
```
